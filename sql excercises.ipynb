```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
database = "sqlite_db_pythonsqlite.db"
connection = sql.connect(database)
```

# Q10:

```python
query = '''
WITH unfiltered AS
(
SELECT f.name as facility,
SUM(
CASE WHEN m.memid = 0 THEN f.guestcost
ELSE f.membercost END) AS total_revenue
FROM Bookings AS b
INNER JOIN Members AS m
USING (memid)
INNER JOIN Facilities AS f
USING (facid)
GROUP BY f.name
)
SELECT *
FROM unfiltered
WHERE total_revenue < 1000
ORDER BY total_revenue
'''
```

```python
revenueSources = pd.read_sql_query(query, connection)
```

```python
revenueSources
```

# Q11:

```python
query = '''
SELECT m1.surname||','||' '||m1.firstname AS member,  m2.surname||','||' '||m2.firstname AS recommended_by
FROM Members AS m1
LEFT JOIN Members AS m2
ON m1.recommendedby = m2.memid
ORDER BY m1.surname, m1.firstname
'''
```

```python
membersPaired = pd.read_sql_query(query, connection)
```

# Q12:

```python
query = '''
SELECT f.name AS facility, COUNT (*) AS usage
FROM Bookings AS b
LEFT JOIN facilities AS f
USING (facid)
WHERE b.memid <> 0
GROUP BY facid
'''
```

```python
usageByMember = pd.read_sql_query(query, connection)
```

```python
usageByMember
```

# Q13:

```python
query = '''
SELECT f.name AS facility, strftime('%m', starttime) AS month, COUNT (*) AS usage
FROM Bookings AS b
LEFT JOIN facilities AS f
USING (facid)
GROUP BY facid, month
'''
```

```python
usageByMonth = pd.read_sql_query(query, connection)
```
