## Introduction<a id='4.2_Introduction'></a>

The goal for this report is to answer two fundamental questions for ski resorts such as Big Mountain based on a detailed inventory of their facilities:

1) How much can you charge for tickets before you start pricing yourself out of the market?

2) Based on what the market will bare, how much more can you charge if you offer the customer greater value by adding to your skiing layout? How much &#0160;&#0160;&#0160;&#0160;less if you go in the opposite direction and downsize?

The revenue impact assumes that there are 350,000 annual visitors buying an average of five tickets a-piece.  One conclusion that stands out here is that you can close one of your runs and have no market impact.

We can also analyze the impact of making several changes at the same time.  For example, if you add a run, increase your vertical drop by 150 feet and raise your total number of lift chairs by 1, you can expect an increase in the ticket price of 1.57 and an annual increase in revenues of 2,750,000.
